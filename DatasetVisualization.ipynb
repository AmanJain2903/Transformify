# DATASET VISUALIZATION

```python
import tensorflow as tf 
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
BATCH_SIZE = 2
```

```python
trainDataset = tf.data.Dataset.load('SavedDataset/trainDataset')
valDataset = tf.data.Dataset.load('SavedDataset/valDataset')
```

```python
print(len(trainDataset)*BATCH_SIZE, len(valDataset)*BATCH_SIZE)
```

```python
#         R        G       B
MEAN = [123.675, 116.28, 103.53]
STD = [58.395, 57.12, 57.375]

# These Values Were Obtained From The ImageNet Dataset
```

```python
plt.figure(figsize=(15, 10))
for data in trainDataset.take(1):
    images, labels = data['pixel_values'], data['labels']
    for i in range(4):
        ax = plt.subplot(1, 4, i+1)
        if i%2==0:
            image = tf.transpose(images[i//2], (1, 2, 0))
            plt.imshow(image)
            plt.title('Image')
        else:
            plt.imshow(labels[i//2])
            plt.title('Label')
        plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15, 10))
for data in valDataset.take(1):
    images, labels = data['pixel_values'], data['labels']
    for i in range(4):
        ax = plt.subplot(1, 4, i+1)
        if i%2==0:
            image = tf.transpose(images[i//2], (1, 2, 0))
            plt.imshow(image)
            plt.title('Image')
        else:
            plt.imshow(labels[i//2])
            plt.title('Label')
        plt.axis('off')
plt.show()
```
